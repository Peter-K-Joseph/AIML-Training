```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

datasheet = pd.read_csv('./datasheet.csv')
dataframe = pd.DataFrame(datasheet)
# Create a bar graph that contains name and salary of all the employees
fig = plt.figure()
plt.subplot(2, 1, 1).bar(dataframe['Name'], dataframe['Salary'])
plt.title('Salary of all employees')
plt.xlabel('Name')
plt.ylabel('Salary')
```

```python
# Line graph for name and salary of all the employees
fig = plt.figure()
plt.subplot(8, 1, 2).plot(dataframe['Name'], dataframe['Salary'])
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary of all employees')
```

```python
# Histogram of salary
fig = plt.figure()
plt.subplot(1, 1, 1).hist(dataframe['Salary'], bins=10)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary of all employees')
```

```python
# scatter plot for name and salary of all the employees
fig = plt.figure()
plt = plt.subplot(2, 1, 1).scatter(dataframe['Name'], dataframe['Salary'])
plt.title('Salary of all employees')
```

```python
# pie plot for salary
fig = plt.figure()
plt = plt.subplot(1, 1, 1).pie(dataframe['Salary'], labels=dataframe['Name'])
```
